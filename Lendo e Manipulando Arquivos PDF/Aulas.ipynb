## Aula 4 - rotaes 

#### Usando o metodo da pagina

```python
# --- Rotacionando

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'vpt_minecraft.pdf'
leitor_pdf = pypdf.PdfReader(caminho_pdf)

escritor_pdf = pypdf.PdfWriter()

for page in leitor_pdf.pages:
    rotated_page = page.rotate(90)
    escritor_pdf.add_page(rotated_page)

escritor_pdf.write(PastaDeDocCriados/'vpt_minecraft_rotacionado_01.pdf')
```

##### Criando um copia do pdf

```python

# --- Agilizando processo: usando argumento clone_from no PdfWriter!

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'vpt_minecraft.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

for page in escritor_pdf.pages:
    page.rotate(-90)

escritor_pdf.write(PastaDeDocCriados/'vpt_minecraft_rotacionado_02.pdf')
```

#### Usando um objeto -> transform --> repetvel

```python
# --- Usando um Transformation para rodar contedo

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'vpt_minecraft.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

transformation = pypdf.Transformation().rotate(45)

for page in escritor_pdf.pages:
    page.add_transformation(transformation)

escritor_pdf.write(PastaDeDocCriados/'vpt_minecraft_rotacionado_03.pdf')

```

##### Aplicando duas transformaes 

```python

# --- Adicionando uma translao ao Transformation para tentar centralizar o contedo

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'vpt_minecraft.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

transformation = pypdf.Transformation().rotate(30).translate(tx=300)

for page in escritor_pdf.pages:
    page.add_transformation(transformation)

escritor_pdf.write(PastaDeDocCriados/'vpt_minecraft_rotacionado_04.pdf')
```

## Aula 5

### Reduzindo proporcionalmente

```python
# --- Redimensionando a pgina - muda apenas o que  considerado 100%

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'vpt_minecraft.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

for page in escritor_pdf.pages:
    page.scale_by(0.1)

escritor_pdf.write(PastaDeDocCriados/'vpt_minecraft_redimensionado_01.pdf') # diminui o tamanho da pgina em para 10% do original
```

### Redimencionando apenas o contedo e no o formato da pgina usando o Transformation

```python

# --- Redimensionando o contedo - muda posio do contedo na pgina

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'vpt_minecraft.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

transformation = pypdf.Transformation().scale(sx=1.5, sy=0.5)

for page in escritor_pdf.pages:
    page.add_transformation(transformation)

escritor_pdf.write(PastaDeDocCriados/'vpt_minecraft_redimensionado_03.pdf')
```

```python
# --- Revertendo processo - contedo continua no PDF!

from pathlib import Path

import pypdf

caminho_pdf = Path(PastaDeDocCriados/'vpt_minecraft_redimensionado_03.pdf')
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

transformation = pypdf.Transformation().scale(sx=0.666, sy=2)

for page in escritor_pdf.pages:
    page.add_transformation(transformation)

escritor_pdf.write(PastaDeDocCriados/'vpt_minecraft_redimensionado_04.pdf')
```

## Aula 7 - Carimbo

```python
# --- Inserindo o carimbo com seu tamanho atual

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'dom_casmurro.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

caminho_carimbo = Path('materiais de aula') / 'assets' / 'carimbo.pdf'
carimbo = pypdf.PdfReader(caminho_carimbo).pages[0]

for pagina in escritor_pdf.pages:
    pagina.merge_page(carimbo, over=True)

escritor_pdf.write(PastaDeDocCriados/'dom_casmurro_carimbo.pdf')
```

```python


# ---  Redimensionando e centralizando o carimbo

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'dom_casmurro.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

caminho_carimbo = Path('materiais de aula') / 'assets' / 'carimbo.pdf'
carimbo = pypdf.PdfReader(caminho_carimbo).pages[0]

for pagina in escritor_pdf.pages:
    fator_escala = 4
    x = pagina.mediabox.left
    y = pagina.mediabox.top - carimbo.mediabox.height / fator_escala
    transformation = pypdf.Transformation().scale(1 / fator_escala).translate(x, y)
    pagina.merge_transformed_page(carimbo, transformation, over=True)

escritor_pdf.write(PastaDeDocCriados/'dom_casmurro_carimbo_ajustado.pdf')
```

```python



# --- Marca d'gua com mesmo tamanho da pgina

from pathlib import Path

import pypdf

caminho_pdf = Path('materiais de aula') / 'documentos' / 'dom_casmurro.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)

caminho_marca = Path('materiais de aula') / 'assets' / 'marca.pdf'
marca = pypdf.PdfReader(caminho_marca).pages[0]

for pagina in escritor_pdf.pages:
    escala_x = pagina.mediabox.width / marca.mediabox.width
    escala_y = pagina.mediabox.height / marca.mediabox.height
    transformation = pypdf.Transformation().scale(sx=escala_x, sy=escala_y)
    pagina.merge_transformed_page(marca, transformation, over=False)

escritor_pdf.write(PastaDeDocCriados/'dom_casmurro_marca.pdf')
```

```python
# --- Pegando senha do PDF via input
from pathlib import Path

import pypdf

caminho_pdf = Path(PastaDeDocCriados/'dom_casmurro_com_senha.pdf')
leitor_pdf = pypdf.PdfReader(caminho_pdf)

if leitor_pdf.is_encrypted:
    result = leitor_pdf.decrypt("senha-secreta")
    if result == 0:
        print('Senha incorreta')
    else:
        print('Senha correta')
        print(f'O PDF tem {len(leitor_pdf.pages)} pginas.')
else:
    print(f'O PDF tem {len(leitor_pdf.pages)} pginas.')
```

```python
# --- Pegando senha do PDF via input
from pathlib import Path

import pypdf

caminho_pdf = Path(PastaDeDocCriados/'dom_casmurro_com_senha.pdf')
leitor_pdf = pypdf.PdfReader(caminho_pdf)

if leitor_pdf.is_encrypted:
    result = leitor_pdf.decrypt("senha-secreta-errada")
    if result == 0:
        print('Senha incorreta')
    else:
        print('Senha correta')
        print(f'O PDF tem {len(leitor_pdf.pages)} pginas.')
else:
    print(f'O PDF tem {len(leitor_pdf.pages)} pginas.')
```

```python
# --- Pegando senha do PDF via input
from pathlib import Path

import pypdf

caminho_pdf = Path(PastaDeDocCriados/'dom_casmurro_com_senha.pdf')
leitor_pdf = pypdf.PdfReader(caminho_pdf)

if leitor_pdf.is_encrypted:
    senha = input('PDF requer senha. Digite a senha do PDF:')
    result = leitor_pdf.decrypt(senha)
    if result == 0:
        print('Senha incorreta')
    else:
        print('Senha correta')
        print(f'O PDF tem {len(leitor_pdf.pages)} pginas.')
else:
    print(f'O PDF tem {len(leitor_pdf.pages)} pginas.')
```

## Aula 9 - Fazendo Anotaes do PDF

```python
# --- Adicionando anotaes a um PDF

from pathlib import Path

import pypdf
from pypdf.annotations import FreeText, Text, Rectangle

caminho_pdf = Path('materiais de aula') / 'documentos' / 'dom_casmurro.pdf'
escritor_pdf = pypdf.PdfWriter(clone_from=caminho_pdf)
```

```python

# Texto livre
texto_livre = FreeText(text="Ol Mundo!\nEsta  uma anotao flutuante!", rect=(400, 550, 550, 600))
escritor_pdf.add_annotation(0, texto_livre)
```

```python

# Texto em anotao
texto = Text(text="Texto da anotao!", rect=(50, 600, 200, 650))
escritor_pdf.add_annotation(1, texto)
```

```python


# Linha reta
linha = Rectangle(rect=(50, 550, 200, 650))
escritor_pdf.add_annotation(2, linha)
```

```python

# Arquivo anexo qualquer (deve ser lido como bytes, modo de leitura "rb")
caminho_imagem = Path('materiais de aula') / 'assets' / 'cachorro.jpg'
with open(caminho_imagem, 'rb') as arquivo:
    dados = arquivo.read()
escritor_pdf.add_attachment('cachorro.jpg', dados)
```

```python

# Gera PDF com todas as anotaes
escritor_pdf.write(PastaDeDocCriados/'dom_casmurro_anotado.pdf')
```
