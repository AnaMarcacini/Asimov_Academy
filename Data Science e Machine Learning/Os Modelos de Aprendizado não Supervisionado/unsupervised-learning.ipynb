```python
kmeans.inertia_ # somatario quadrado de todas as distancias dos pontos até seus centroides
```

```python
f, ax = plt.subplots(figsize=(25, 6))

list_inertia = []
list_k = []
for k in [2, 3, 4, 5, 6]:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data[0])
    
    list_inertia += [kmeans.inertia_]
    list_k += [k]
ax.set_title("Inercia x quantidade de Clusters (centroides)")
ax.plot(list_k, list_inertia)# cotovelo nesse caso o 4 é o ponto otimo - menor redução de um ponto ao outro
```

### Gaussian Mixture Models

```python
from sklearn.datasets import make_gaussian_quantiles
```

```python
X1, Y1 = make_gaussian_quantiles(n_features=2, n_classes=3)# criando dados
plt.scatter(X1[:, 0], X1[:, 1], marker="o", c=Y1, s=50, edgecolor="k")
# plt.scatter(X1[:, 0], X1[:, 1], marker="x", c=Y1, s=50, edgecolor="k")
```

```python
from sklearn.mixture import GaussianMixture

gm = GaussianMixture(n_components=3)
gm.fit(X1)
```

```python
gm.weights_
```

```python
gm.means_
```

```python
gm.covariances_
```

```python
gm.covariances_.shape
```

```python
from matplotlib.patches import Ellipse

def draw_ellipse(position, covariance, **kwargs):#, **kwargs -- transparencia
    ax = plt.gca()
    
    # Convert covariance to principal axes
    U, s, Vt = np.linalg.svd(covariance)
    angle = np.degrees(np.arctan2(U[1, 0], U[0, 0]))
    width, height = 2 * np.sqrt(s)
    
    # Draw the Ellipse
    for nsig in range(1, 4):
        ax.add_patch(Ellipse(position, nsig * width, nsig * height,
                             angle, **kwargs))
```

```python
def plot_gmm(gmm, X):
    fig, ax = plt.subplots(figsize=(20, 8))
    
    labels=gmm.predict(X)
    ax.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', zorder=2)
    
    w_factor = 0.2/gmm.weights_.max()
    for pos, covar, w, c in zip(gmm.means_, gmm.covariances_, gmm.weights_, ["red", "blue", "green"]):
        draw_ellipse(pos, covar, alpha=w * w_factor, color=c)
```

```python
plot_gmm(gm, X1)
```

### Detecção de anomalias e geração de novas estimativas

```python
gm.sample(10)#amostra de 10 valores
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

labels=gm.predict(X1)
ax[0].scatter(X1[:, 0], X1[:, 1], c=labels, s=40, cmap='viridis')
ax[0].set_title("Conjunto de dados original")

X_new, Y_new = gm.sample(100)
ax[1].scatter(X_new[:, 0], X_new[:, 1], c=Y_new, s=40, cmap='viridis')
ax[1].set_title("conjuntos de dados forjados pelo modelo")
```

passo pontos para o meu grafico e devolve o logaritmo da probabilidade de distribuição acumulado nesse ponto

A densidade da minha função || a altura de cada ponto no eixo z

```python
densities = gm.score_samples(X1)
densities_threshold = np.percentile(densities, 4)# ordenar do menor para o maior - passando 4%
densities_threshold # alguma zona a baixo desses 4% é um ponto de anomialia
```

```python
densities = gm.score_samples(X1)
densities_threshold = np.percentile(densities, 10)# ordenar do menor para o maior - passando 10% 
anomalies = X1[densities < densities_threshold]

fig, ax = plt.subplots(figsize=(15, 5))
ax.scatter(X1[:, 0], X1[:, 1], c=labels, s=40, cmap='viridis')
ax.scatter(anomalies[:, 0], anomalies[:, 1], c="red", s=60, marker="x", cmap='viridis')
```
