### O treinamento de um modelo de regressão linear

```python
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x + np.random.randn(100, 1)# adição de um pequeno valor em x para gerar um ruido || simular um cenario real
y_semRuido = 4+3*x
```

```python
plt.scatter(x,y_semRuido)
```

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.scatter(x, y)
```

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.plot(x, y_semRuido)
plt.scatter(x, y)
```

```python
x
```

```python
x_b = np.c_[np.ones((100, 1)), x]# transformando em uma matriz com duas colunas
```

```python
x_b# transformando em uma matriz com duas colunas
```

```python
theta = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y) # calculando os coeficientes da reta - solução analitica
```

```python
theta# valores de a e b de ax+b=y
```

```python
y_hat = theta[0] + x * theta[1]
```

```python
fig, ax = plt.subplots(figsize=(20, 7))
plt.plot(x, y_hat)
plt.scatter(x, y)
```

### Regressão linear no sklearn

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x, y)
lin_reg.intercept_, lin_reg.coef_
```

```python
eta = 0.1  # learning rate
n_iterations = 10
m = 100

theta = np.random.randn(2,1)  # random initialization -- criando valores aleatorios para o a e b da reta
fig, ax = plt.subplots(figsize=(20, 7))
plt.scatter(x, y)

for iteration in range(n_iterations):
    gradients = 2/m * x_b.T.dot(x_b.dot(theta) - y) # calculo do gradiente
    theta = theta - eta * gradients
    
    if iteration % 1 == 0:
        y_hat = theta[0] + x * theta[1]
        ax.plot(x, y_hat, alpha=(0.1+iteration / n_iterations))
```

### Regressão polinomial

```python
m = 100
x = 6 * np.random.rand(m, 1) - 3 # gernado valores de -3 a 3
y = 0.5 * x**2 + x + 2 + np.random.randn(m, 1)
plt.plot(x)
```

```python
plt.scatter(x, y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)# grau do polinomio = 2
x_poly = poly_features.fit_transform(x) # faço uma matriz com uma coluna o valor de x e a outra seu valor quadratico
```

```python
x_poly[0]
```

```python
x_poly
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_poly, y)
lin_reg.intercept_, lin_reg.coef_
```

```python

lin_reg.intercept_# valor onde ele acha que intercepta o eixo y (coeficiente b)
```

```python

lin_reg.coef_ # estimativa dos coeficientes
```

```python
plt.scatter(x, y)
plt.scatter(x, lin_reg.predict(x_poly))
```

### O que acontece se a complexidade do modelo for aumentada desnecessariamente?

```python
fig, ax = plt.subplots(figsize=(25, 8))
for degree in [1, 2, 300]:
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    x_poly = poly_features.fit_transform(x)
    x_seq = np.linspace(x.min(), x.max(), 300).reshape(-1,1)
    
    lin_reg = LinearRegression()
    lin_reg.fit(x_poly, y)  
    
    plt.scatter(x, y)
    x_seq_transf = poly_features.fit_transform(x_seq)
    plt.plot(x_seq, lin_reg.predict(x_seq_transf), label=degree)
ax.set_ylim([-5, 12])
ax.set_xlim([-4, 4])
```

```python
fig, ax = plt.subplots(figsize=(25, 8))
for degree in [1, 2, 10]:
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    x_poly = poly_features.fit_transform(x)
    x_seq = np.linspace(x.min(), x.max(), 300).reshape(-1,1)
    
    lin_reg = LinearRegression()
    lin_reg.fit(x_poly, y)  
    
    plt.scatter(x, y)
    x_seq_transf = poly_features.fit_transform(x_seq)
    plt.plot(x_seq, lin_reg.predict(x_seq_transf), label=degree)
ax.set_ylim([-5, 12])
ax.set_xlim([-4, 4])
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
m = 100
x = np.random.rand(m, 1) - 3
y = np.random.randn(m, 1)
```

```python
plt.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(25, 8))
plt.scatter(x, y)

for alpha in [0, 1e-5, 1]: # pequena penalização media e grande - quanro maior mais estavel fica o grafico e quanto menor mais erratica
    poly_features = PolynomialFeatures(degree=10, include_bias=False)
    x_poly = poly_features.fit_transform(x)
    x_seq = np.linspace(x.min(), x.max(), 300).reshape(-1,1)# layout grafico
    
    ridge_reg = Ridge(alpha=alpha, solver="cholesky") # metodo de resolução padrão solver="cholesky"
    ridge_reg.fit(x_poly, y)  
    
    
    x_seq_transf = poly_features.fit_transform(x_seq)
    plt.plot(x_seq, ridge_reg.predict(x_seq_transf), label=alpha)
ax.legend()

```

```python
from sklearn import datasets
iris = datasets.load_iris()
list(iris.keys())
```

```python
x = iris["data"][:, 3:]  # petal width - pegar todas as linhas e apenas a terceira coluna (contando de 0-3)
y = (iris["target"] == 2).astype(np.int64)
# transformar em uma lista de true ou false
```

```python
(iris["target"] == 2)
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(x, y)
```

```python
X_new = np.linspace(0, 3, 1000).reshape(-1, 1) ## criar vetor com varios valores (1000) de petalas diferentes (entre 0 e 3)
y_proba = log_reg.predict_proba(X_new) # ver a probabilidade do valor pertencer ou não a classe que criamos
y_proba
```

```python
X_new
```

```python
fig, ax = plt.subplots(figsize=(25, 8))
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris virginica") # todas as linhas que ele não acha que é
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris virginica")# todas as linhas que ele acha que é
```

### Softmax Regression

```python
x = iris["data"][:, (2, 3)]  # petal length, petal width
y = iris["target"]

softmax_reg = LogisticRegression(multi_class="multinomial",solver="lbfgs", C=10)
softmax_reg.fit(x, y)
```

```python
0.5 * np.log(1/0.5) + 0.5 * np.log(1/0.5)
```

```python
softmax_reg.predict([[5, 2]])
softmax_reg.predict_proba([[5, 2]])
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = x[:, 0].min() - 0.5, x[:, 0].max() + 0.5
y_min, y_max = x[:, 1].min() - 0.5, x[:, 1].max() + 0.5
h = 0.02  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = softmax_reg.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(25, 9))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(x[:, 0], x[:, 1], c=y, edgecolors="k", cmap=plt.cm.Paired)
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

plt.show()
```
